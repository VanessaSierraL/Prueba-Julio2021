```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set()
```

```python
e = np.random.randn(30)
s = np.zeros(30)

s[10] = 1000

for i in range(11, 30):
    s[i] = 0.5 * s[i-1] + e[i]

ts = pd.DataFrame(s, 
                  index=pd.date_range("2020-01-01", periods=30),
                  columns=['x1'])
ts.plot()
```
